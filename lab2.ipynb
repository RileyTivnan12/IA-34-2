# D. Group and Visualize the Data

```python
import matplotlib.pyplot as plt

dept_counts = df.groupby("department")["id"].count()

time_counts = df.groupby("data_as_of")["id"].count()

# 1: Bar chart of tools by department
plt.figure(figsize=(6,4))
dept_counts.plot(kind="bar")
plt.title("Number of AI Tools by Department")
plt.xlabel("Department")
plt.ylabel("Count of Tools")
plt.xticks(rotation=45)
plt.show()

# 2: Line chart of tool counts by 'data_as_of'
plt.figure(figsize=(6,4))
time_counts.plot(kind="line", marker="o")
plt.title("AI Tool Entries Over Time")
plt.xlabel("Data As Of")
plt.ylabel("Count of Tools")
plt.grid(True)
plt.show()
```

- Visualization 1 = Bar chart of tools by department
The bar chart shows that most AI tools were reported by DPH, which SFPUC and a few other agencies then followed.
From this visualization, it is suggesting that health-related applications are currently the largest area of AI adoption.

- Visualization 2 = Line chart of entries over time
The line chart shows tool reporting clustered around a few key dates, with most entries sharing the same "as of" date.
This reflects bulk submissions rather than continuous reports.
